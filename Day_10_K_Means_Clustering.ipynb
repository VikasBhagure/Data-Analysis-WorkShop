```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/income.csv")
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data['Age'] = scaler.fit_transform(data[['Age']])
data['Income($)'] = scaler.fit_transform(data[['Income($)']])
```

```python
plt.scatter(data['Age'],data['Income($)'],s=100)
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

```python
x = data[['Age','Income($)']]
x
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
ypred = km.fit_predict(x)
ypred
```

```python
x['Cluster'] = ypred
x
```

```python
centroids = km.cluster_centers_
```

```python
df1 = x[x['Cluster']==0]
df2 = x[x['Cluster']==1]
df3 = x[x['Cluster']==2]
plt.scatter(df1['Age'],df1['Income($)'],color='green',label="Cluster 1",s=150)
plt.scatter(df2['Age'],df2['Income($)'],color='red',label="Cluster 2",s=150)
plt.scatter(df3['Age'],df3['Income($)'],color='blue',label="Cluster 3",s=150)
plt.scatter(centroids[:,0],centroids[:,1],s=200,marker="*",label="Centroid",color='purple')
plt.show()
```

```python
#### Elbow Method
sse = []
for k in range(1,20):
  km = KMeans(n_clusters=k)
  km.fit(data[['Age','Income($)']])
  sse.append(km.inertia_)
```

```python
plt.xlabel("K Value")
plt.ylabel("SSE")
plt.xticks(range(1,20))
plt.plot(range(1,20),sse)
plt.show()
```
