```python
#### Remove Null Values
data.dropna(axis=0)   ## drops rows having missing values
```

```python
data.dropna(axis=1)   ## drops columns having missing values
```

```python
data.dropna(axis=0,how='all')   ## drop rows which are completely null
```

```python
data.dropna(axis=0,how='any')   ## drop rows where any value is null 
```

```python
data.dropna(axis=0,how='any',subset=['Team'])
```

```python
### Fill Null Values
```

```python
data.isnull().sum()
```

```python
data['First Name'] = data['First Name'].fillna('Unknown')
```

```python
mode_gender = data['Gender'].mode()
mode_sm = data['Senior Management'].mode()
mode_team = data['Team'].mode()
mode_team
```

```python
data['Gender'] = data['Gender'].fillna('Female')
```

```python
data['Senior Management'] = data['Senior Management'].fillna('True')
data['Team'] = data['Team'].fillna('Client Services')
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data.info()
```

```python
### Change Datatype of Columns
```

```python
data['Start Date'] = data['Start Date'].astype('datetime64[ns]')
data['Last Login Time'] = data['Last Login Time'].astype('datetime64[ns]')
data['Senior Management'] = data['Senior Management'].astype('bool')
```
